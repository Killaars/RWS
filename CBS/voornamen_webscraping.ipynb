```python
from bs4 import BeautifulSoup
import urllib.request
fp = urllib.request.urlopen('https://www.meertens.knaw.nl/nvb/naam/is/lars')
soup = BeautifulSoup(fp, 'html.parser')
print(soup)
```

```python
import numpy as np
x=soup.find_all("script")[5].string.splitlines()[3]
freq = [float(y) for y in x.split('(')[1].split(')')[0].split(',')]
print(freq)

x=soup.find_all("script")[5].string.splitlines()[2]
year = [int(y) for y in x.split('(')[1].split(')')[0].split(',')]
print(year)
```

```python
import pandas as pd

df = pd.DataFrame({"year":year,"count":freq})
df["leeftijd"] = 2018-np.array(year)
sum_count = df["count"].sum()
df["kans"] = df["count"].apply(lambda x: x/sum_count)
gewogen_leeftijd = (np.array(df["leeftijd"]) * np.array(df["kans"])).sum()

print(gewogen_leeftijd)
```

```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
import numpy as np

names = ['lars','steven','maikel','wessel','marloes','max','jeroen','vincent','jordi','alexander','elena','frederique','laura','josine','pauline','jose','siemen','mariel','maarten','sjoerd','heidi','sebastian','laurent','bart','piet','marco','bas']
leeftijd=[]
for name in names:
    print(name)
    fp = urllib.request.urlopen('https://www.meertens.knaw.nl/nvb/naam/is/%s' %(name))
    soup = BeautifulSoup(fp, 'html.parser')
    x=soup.find_all("script")[5].string.splitlines()[3]
    freq = [float(y) for y in x.split('(')[1].split(')')[0].split(',')]

    x=soup.find_all("script")[5].string.splitlines()[2]
    year = [int(y) for y in x.split('(')[1].split(')')[0].split(',')]
    
    df = pd.DataFrame({"year":year,"count":freq})
    df["leeftijd"] = 2018-np.array(year)
    sum_count = df["count"].sum()
    df["kans"] = df["count"].apply(lambda x: x/sum_count)
    gewogen_leeftijd = (np.array(df["leeftijd"]) * np.array(df["kans"])).sum()
    
    leeftijd = np.append(leeftijd,gewogen_leeftijd)
    print(gewogen_leeftijd)
```

```python
print(leeftijd)
print(np.mean(leeftijd))
print(np.std(leeftijd))
```
